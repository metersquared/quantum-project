Creating QuantumVariable

```python
from qrisp import QuantumVariable, h, z, cx
qv = QuantumVariable(5)

h(qv[0])
z(qv)
cx(qv[0], qv[3])

print(qv.qs)
print(qv)
```

Abstraction of QuantumVariable

```python
from qrisp import QuantumChar
qch = QuantumChar()

qch[:] = "e"

cx(qv[0], qch[0])
print(qch.qs)
```

Get Measurements

```python
results = qch.get_measurement()
print(results)
```
